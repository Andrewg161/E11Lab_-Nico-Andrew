```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
outside_data = pd.read_csv('wk5_outside_data.csv')
outside_data
```

```python
restroom_data = pd.read_csv('wk5_restroom_data.csv')
restroom_data
```

```python
#temp
plt.plot(restroom_data["Temperature (C)"].iloc[1:300], label="Restroom Data")
plt.plot(outside_data["Temperature (C)"].iloc[1:300], label="Outside Data", color="pink")
plt.ylabel('Temperature (C)')  
plt.xlabel('Time (s)')  
plt.title("Time Series")  
plt.legend()
plt.show()

print("Average inside temp (C):",np.mean(restroom_data["Temperature (C)"].iloc[1:300]))
print("Average outside temp (C):",np.mean(outside_data["Temperature (C)"].iloc[1:300]))
print("STD inside temp (C):",np.std(restroom_data["Temperature (C)"].iloc[1:300]))
print("STD outside temp (C):",np.std(outside_data["Temperature (C)"].iloc[1:300]))
print("Mean Uncertainty inside temp (C):",(np.std(restroom_data["Temperature (C)"].iloc[1:300]))/np.sqrt(299))
print("Mean Uncertainty inside temp (C):",(np.std(outside_data["Temperature (C)"].iloc[1:300]))/np.sqrt(299))
```

```python
#temp
plt.hist(restroom_data["Temperature (C)"].iloc[1:300], bins=8, label="Restroom Data")
plt.hist(outside_data["Temperature (C)"].iloc[1:300], bins=8, label="Outside Data", color="pink", alpha=0.9) 
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')  
plt.title("Histogram")
plt.legend()
plt.show()

outside_mean = np.mean(outside_data["Temperature (C)"].iloc[1:300])
std_outside = np.std(outside_data["Temperature (C)"].iloc[1:300])
x = np.linspace(min(outside_data["Temperature (C)"].iloc[1:300]), max(outside_data["Temperature (C)"].iloc[1:300]))
y = norm.pdf(x, outside_mean, std_outside)
plt.plot(x, y, label="Normal Distribution")
restroom_mean = np.mean(restroom_data["Temperature (C)"].iloc[1:300])
std_restroom = np.std(restroom_data["Temperature (C)"].iloc[1:300])
x = np.linspace(min(restroom_data["Temperature (C)"].iloc[1:300]), max(restroom_data["Temperature (C)"].iloc[1:300]))
y = norm.pdf(x, restroom_mean, std_restroom)
plt.plot(x, y, label="Normal Distribution")

print("Average restroom temp (C):",np.mean(restroom_data["Temperature (C)"].iloc[1:300]))
print("Average outside temp (C):",np.mean(outside_data["Temperature (C)"].iloc[1:300]))
print("STD restroom temp (C):",np.std(restroom_data["Temperature (C)"].iloc[1:300]))
print("STD outside temp (C):",np.std(outside_data["Temperature (C)"].iloc[1:300]))
print("Mean Uncertainty restroom Temperature (C):",(np.std(restroom_data["Temperature (C)"].iloc[1:300]))/np.sqrt(299))
print("Mean Uncertainty outside Temperature (C):",(np.std(outside_data["Temperature (C)"].iloc[1:300]))/np.sqrt(299))
```
