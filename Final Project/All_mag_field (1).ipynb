```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

files = [
    "magnetometer_1.csv",
    "magnetometer_2.csv",
    "magnetometer_3.csv",
    "magnetometer_4.csv",
    "magnetometer_5.csv",
    "magnetometer_6.csv"
]

site_names = [
    "Etcheverry Hall Basement",
    "Berkeley QB3",
    "Memorial Glade",
    "RSF",
    "Downtown Berkeley BART",
    "Cesar Chavez Park"
]

plt.figure(figsize=(12, 6))

average_fields = []

for file, label in zip(files, site_names):
    df = pd.read_csv(file)
    
    if "Mag_Total (uT)" in df.columns:
        mag = df["Mag_Total (uT)"]
    else:
        mag = df.iloc[:, -1]
    
    avg_mag = mag.mean()
    average_fields.append((label, avg_mag))
    
    time = np.arange(len(mag))
    plt.plot(time, mag, label=label)

plt.title("Magnetic Field Strength Over Time (All Spots)")
plt.xlabel("Time in Seconds")
plt.ylabel("Magnetic Field Total (μT)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("Average Magnetic Field (μT) at Each Location:")
for site, avg in average_fields:
    print(f"{site}: {avg:.2f} μT")
```
